# NiB    (수정할 텍스트)
### 다양한 종류의 신경 영상 파일을 읽고 쓸 수 있게 해주는 도구
-> 신경 영상 데이터 입출력을 위한 NumPy

### 1. 핵심기능
디스크에 있는 뇌 영상 파일을 파이썬 환경 내에서 사용 가능한 형태로 가져오고, 데이터를 파일로 저장하는 역할을 수행.

- **데이터 읽기** : NIfTI, ANALYZE 등 복잡한 신경 영상 파일을 '이미지 객체'로 불러옴
- **데이터 쓰기** : NumPy 배열 형태의 영상 데이터를 표준 신경 영상 파일 형식으로 저장
- **메타데이터 접근** : 신경 영상 파일은 단순히 픽셀의 집합이 아님. NiBabel을 통해 헤더 정보에 완전히 접근 가능
- **공간 정보 처리** : 데이터의 3D배열이 물리적 공간에 어떻게 위치해 있는지 알려주는 아핀 행렬(Affine matrix)을 올바르게 처리

### 2. 주요 개념

##### 2-1 : 이미지 객체 (Image Object)
`img = nib.load('my_brain.nii.gz')`처럼 파일을 불러올 때 얻게 되는 메인 객체. 메모리 효율성을 위해 실제 이미지 데이터는 사용자가 요청하기 전까지는 메모리에 불러오지 않음.

##### 2-2 : 이미지 데이터 배열 (Image Data Arrary)
`img.get_fdata()`를 통해 접근하는 실제 이미지 데이터.
- 데이터는 NumPy 배열 형태로 반환됨
- 배열의 차원은 영상의 차원과 일치함

##### 2-3 : 아핀 행렬 (Affine Arrary)
NumPy 배열의 복셀 좌표를 실제 세계의 스캐너 좌표로 변환해주는 4x4 행렬.
`img.affine`을 통해 접근.
![alt text](image.png)

- (i, j, k) : 영상 데이터 배열에서의 복셀 인덱스
- (x_real, y_real, z_real) : 실제 환자 공간에서의 물리적 좌표 (mm)
- 좌측 상단 3x3 행렬 : 복셀의 회전(rotation) 및 스케일링(scaling) 정보를 담고 있음.
- 우측 상당 3x1 열벡터 : 이동(translation) 정보를 나타내며, 영상의 원점(0, 0, 0)이 실제 공간의 어디에 위치하는지 정의함

**아핀 행렬의 역할 및 중요성**
- 공간 정합 : 서로 다른 조건에서 촬용된 영상들을 동일한 좌표공간에 정렬시킬때 사용
- 좌표 변환 : 영상의 즉정 복셀 위치가 뇌의 어느 해부학적 위치에 해당하는지 파악
- 데이터 시각화 : 복셀 데이터를 2D 슬라이스나 3D 모델로 정확하게 시각화하
